```python
from llama_stack_client import LlamaStackClient
import pandas as pd
import json
import time
```

```python
client = LlamaStackClient(
    base_url="http://lls-route-model-namespace.apps.rosa.trustyai-rob.4osv.p3.openshiftapps.com"
)
```

```python
benchmarks = client.alpha.benchmarks.list()
print("\n=== Available Benchmarks ===")
for benchmark in benchmarks:
    print(benchmark.identifier)
```

```python
job = client.alpha.eval.run_eval(
    benchmark_id="trustyai_garak::standard",
    benchmark_config={
        "eval_candidate": {
            "type": "model",
            "model": "vllm/qwen3",
            "sampling_params": {
                "max_tokens": 100
            },
        }
     },
)
```

```python
poll_interval = 20
pre_defined_benchmark_id = "trustyai_garak::standard"

def get_job_status(job_id, benchmark_id):
    return client.alpha.eval.jobs.status(job_id=job_id, benchmark_id=benchmark_id)

while True:
    job = get_job_status(job_id=job.job_id, benchmark_id=pre_defined_benchmark_id)
    print(job)

    if job.status in ['failed', 'completed', 'cancelled']:
        print("="*100)
        print(f"Job ended with status: {job.status}")
        break

    time.sleep(poll_interval)
```
